```python
import pandas as pd 
import numpy as np
```

```python
data = {'Name': ['Akash', 'Arun', 'Payal', 'Nithin', 'Ravi', 'Pooja', 'Riya'], 'Age': [18, 17, 19, 17, 18, 19, 19], 'Gender': ['M', 'M', 'F', 'M', 'M', 'F', 'F'], 'Marks': [93, 85, 'NaN', 84, 55, 'NaN', 61]} 
df = pd.DataFrame(data) 
df
```

```python
count=avg = 0
for ele in df['Marks']:    
     if str(ele).isnumeric():        
        count=count+1        
        avg = avg+ele
        avg = avg/count 
```

```python
df = df.replace(to_replace="NaN",value=avg) 
df
```

```python
df['Gender'] = df['Gender'].map({'M': 0,'F': 1, }).astype(float)
df
```

```python
df = df[df['Marks'] >= 75].copy() 
df.drop('Age', axis=1, inplace=True) 
df
```
