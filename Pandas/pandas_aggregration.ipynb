```python
data1 = {'Name':['Akash', 'Arun', 'Priya', 'Payal'], 'Age':[20, 18, 21, 20], 'Address':['Chennai', 'Delhi', 'Bengaluru', 'Mumbai'], 'Qualification':['B.Tech', 'MTech', 'MCA', 'PhD']}    
```

```python
data2 = {'Name':['Abhishek', 'Arun', 'Hitesh', 'Payal'], 'Age':[19, 18, 21, 20], 'Address':['Hyderabad', 'Delhi', 'Punjab', 'Mumbai'], 'Qualification':['Btech', 'MTech', 'Bcom', 'PhD']}
```

```python
df = pd.DataFrame(data1,index=[0, 1, 2, 3]) 
```

```python
df1 = pd.DataFrame(data2, index=[4, 5, 6, 7]) 
print(df, "\n\n", df1) 
```

```python
result = pd.concat([df, df1])
result
```

```python
result = pd.concat([df, df1], axis=1, join='inner')
result
```

```python
result = pd.concat([df, df1], axis=1, join='outer')
result
```

```python
result = pd.concat([df, df1], axis=1)
result = result.reindex(df.index)
result
```

```python
df = pd.DataFrame(data1,index=[0, 1, 2, 3]) 
df1 = pd.DataFrame(data2, index=[4, 1, 5, 3]) 
print(df, "\n\n", df1) 
```

```python
res = pd.concat([df, df1], ignore_index=True)
res
```

```python
df = pd.DataFrame(data1,index=[0, 1, 2, 3]) 
df1 = pd.DataFrame(data2, index=[4, 1, 5, 3]) 
print(df, "\n\n", df1) 
```

```python
res = pd.concat([df, df1], keys=['x', 'y'])
res
```

```python
result = pd.concat([df, df1], axis=1, join='inner')
result
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'], 'Name':['Akash', 'Arun', 'Gaurav', 'Pooja'],         'Age':[21, 20, 22, 24],}    
data2 = {'key': ['K0', 'K1', 'K2', 'K3'], 'Address':['Delhi', 'Chennai', 'Delhi', 'Mumbai'],         'Qualification':['Btech', 'MTech', 'Bcom', 'BSc']}   
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)    
print(df, "\n\n", df1) 
```

```python
result = pd.merge(df, df1, on='key')
result
```

```python
df = pd.DataFrame(data1) 
df1 = pd.DataFrame(data2)    
print(df, "\n\n", df1) 
```

```python
result = pd.merge(df, df1, how='left', on=['key'])
result 
```

```python
result = pd.merge(df, df1, how='right', on=['key'])
result
```

```python
result = pd.merge(df, df1, how='inner', on=['key'])
result
```
